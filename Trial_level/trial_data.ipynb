```python
def prep_data():
    
    gaze_df = pd.read_csv('Raw_data/sel-att-data-185-188-May26-2023.csv')
    gaze_df['subID'] = gaze_df['subID'].astype('string')

    gaze_df['subject'] = gaze_df['subID'].str[-2:]
    gaze_df['subID'] = gaze_df['subID'].astype('string')
    
    gaze_df.sort_values(by=['subject','trial'],inplace=True)
    gaze_df.pop('subject')

    return gaze_df
```

```python
def get_trial(*args):
    gaze_trial = prep_data()
        
    columns = ['subID','expID','trial','pretrained','learned']
    
    for arg in args:
        try:
            gaze_trial[arg]
        except:
            print(f'no column named \"{arg}\"')
        else:
            if arg not in columns:
                columns.append(arg)
                
    gaze_trial.reset_index(inplace=True,drop=True)
    
    return gaze_trial[columns]
```
