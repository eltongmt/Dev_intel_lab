```python
import pandas as pd
import numpy as np
import os
```

```python
gaze_df = pd.read_csv('Raw_data/sel-att-data-185-188-May26-2023.csv')
gaze_df
```

```python
trial_group = ['subID','expID','trial']
gaze_df['pretrained'] = gaze_df['pretrained'].apply(lambda x: 1 if x > 0 else 0)

gaze_df_T = gaze_df.groupby(by=trial_group).mean()
gaze_df_T.pop('pretrained')
group = gaze_df.groupby(by=trial_group).sum()
learned = gaze_df.groupby(by=['subID','expID']).sum()

total_learned = (learned['learned'] // 6).to_list()
learned_per_trial = group['learned'].to_list()
pretrained_per_trial = group['pretrained'].to_list()
```

```python
gaze_df
```

```python
### BROADCASTING WORDS LEARNED 165 t0 6561

n_learned = [0 for i in range(6561)]
count = 0
j=0

for i in range(len(n_learned)):
    if count == 27:
        j+= 1
        count = 0
    
    count += 1
    n_learned[i] = total_learned[j]
            
    
lb  = np.array(['1','2','3','4'])
lb = np.resize(lb,(26244,1))

gaze_df['labelling_window'] = lb
```

```python
len(n_learned)
```

```python
## Primary template is done 
test = gaze_df_T.reset_index()
#test.pop('pretrained')
test.pop('learned')
test.insert(1,'total_learned',n_learned)
test.insert(4,'pretrained',pretrained_per_trial)
test.insert(5,'trial_learned',learned_per_trial)
info = ['subID','total_learned','expID','trial','pretrained','trial_learned']
```

```python
# VALUES FOR ONSET
values = test.columns.to_list()[6:]


pivot = gaze_df.pivot(index=['subID','trial'],columns='labelling_window',values=values)
pivot.columns = pivot.columns.map('_'.join).str.strip('_')
a = pivot.reset_index()


test = pd.merge(test[info + values], a.iloc[:,2:], left_index=True, right_index=True)
```
